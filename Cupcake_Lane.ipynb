```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate yearly production
def calculate_production(production_rate, cooking_time, capacity, yearly_goal):
    total_per_day = min(production_rate * (24 * 60) / cooking_time, capacity)
    yearly_production = total_per_day * 365
    print(f"\nüéØ Initial Goal: {yearly_goal} cupcakes/year")
    print(f"üè≠ Current Capacity: {capacity} cupcakes per cycle")
    print(f"üîÑ Calculated Yearly Production: {int(yearly_production)} cupcakes/year")

    if yearly_production >= yearly_goal:
        print("‚úÖ Your system can meet or exceed the goal!")
    else:
        print("‚ö†Ô∏è Your system falls short of the goal. Consider increasing the oven capacity or production rate.")

    return yearly_production

# Interactive user input
def user_input():
    print("Welcome to the Cupcake Production Simulator! üéÇ\nLet's set up your production line:")
    production_rate = int(input("Enter your cupcake production rate (cupcakes per minute): "))
    cooking_time = int(input("Enter cooking time per batch (in minutes): "))
    capacity = int(input("Enter oven capacity (maximum cupcakes per batch): "))
    yearly_goal = int(input("Enter your desired yearly production goal (in cupcakes): "))

    return production_rate, cooking_time, capacity, yearly_goal

# Project production over 5 years
def project_5_years(initial_production):
    growth_rate = 0.05  # Assume a 5% yearly increase in efficiency
    years = np.arange(1, 6)
    production = [initial_production * (1 + growth_rate) ** year for year in years]
    
    plt.figure(figsize=(10, 6))
    plt.plot(years, production, marker='o', linestyle='--')
    plt.title("Cupcake Production Projection (5 Years)")
    plt.xlabel("Years")
    plt.ylabel("Total Cupcakes Produced")
    plt.grid(True)
    plt.show()

def main():
    production_rate, cooking_time, capacity, yearly_goal = user_input()
    initial_production = calculate_production(production_rate, cooking_time, capacity, yearly_goal)
    
    print("\nüìà Projecting production over the next 5 years with a 5% efficiency increase per year...\n")
    project_5_years(initial_production)

if __name__ == "__main__":
    main()
```
